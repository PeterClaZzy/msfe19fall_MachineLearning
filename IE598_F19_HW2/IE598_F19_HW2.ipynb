```python
import pandas as pd
import numpy as np
treasury = pd.read_csv("D:/UIUC_Course/Machine Learning/Dataset//Treasury Squeeze test - DS1.csv",header = None)
treasury = np.array(treasury)
treasury = np.delete(treasury, 0, axis = 0)
treasury = treasury[:,2:12]
treasury_data=treasury[:,0:9]
treasury_target=treasury[:,9]
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(treasury_data, treasury_target, test_size=0.3, random_state=1, stratify=treasury_target)
k_range = range(1, 26)
scores = []

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
#print(scores)
for k in k_range:
    print ("k =",k,' ',"Accuracy =",scores[k-1]*100,"%")

import matplotlib.pyplot as plt
plt.title('KNN with different k')
plt.plot(k_range,scores)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn import tree

TreasuaryTreeModel = tree.DecisionTreeClassifier(criterion= 'gini',max_depth=3,random_state=1)
TreasuaryTreeModel.fit(X_train, y_train)

y_predict = TreasuaryTreeModel.predict(X_test)
print(accuracy_score(y_test, y_predict))

tree.plot_tree(TreasuaryTreeModel)

print("My name is Jialing Zhu")
print("My NetID is: jz72")
print("I hereby certify that I have read the University policy on Academic Integrity and that I am not in violation.")
```
